## Image Cropping from Object Type Bounding Box

```python
# Directories
image_dir = "dataset/images"
json_dir = "dataset/masks"
output_dir = "dataset/croppedImgs"

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Process each image
for image_file in os.listdir(image_dir):
    if not image_file.endswith(".png"):
        continue

    image_path = os.path.join(image_dir, image_file)
    json_path = os.path.join(json_dir, os.path.splitext(image_file)[0] + ".json")

    if not os.path.exists(json_path):
        print(f"JSON file not found for {image_file}")
        continue

    # Load image
    image = Image.open(image_path)

    # Load corresponding JSON file
    with open(json_path, "r") as f:
        data = json.load(f)

    # Extract and crop each rectangle
    for i, shape in enumerate(data.get("shapes", [])):
        if shape["shape_type"] != "rectangle":
            continue

        (x1, y1), (x2, y2) = shape["points"]
        x1, y1, x2, y2 = map(int, [x1, y1, x2, y2])
        cropped = image.crop((x1, y1, x2, y2))

        # Save cropped image
        base_name = os.path.splitext(image_file)[0]
        cropped_filename = f"{base_name}.png"
        cropped.save(os.path.join(output_dir, cropped_filename))

print("Cropping complete.")
```
